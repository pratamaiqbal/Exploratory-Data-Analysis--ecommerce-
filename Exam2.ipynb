```python
df.shape
```

Data ini memiliki 609 data pertama kita akan cek apakah terdapat data NULL atau tidak.

```python
df.isna().sum()
```

Untuk original price yang kosong kita akan isi dengan rata-rata dari original price

```python
df['Original_price'].fillna(df['Original_price'].mean(),inplace = True)
```

```python
df.isna().sum()
```

Untuk perc_of_unit_sold yang kosong kita akan isi dengan rata-rata dari perc_of_unit_sold

```python
df['perc_of_unit_sold'].fillna(df['perc_of_unit_sold'].mean(),inplace=True)
```

```python
df.isna().sum()
```

Untuk rating yang kosong kita akan isi dengan rata-rata dari rating

```python
df['rating'].fillna(df['rating'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
df['shpping_rate'].mode()
```

```python
df['gender'].mode()
```

Untuk shpping_rate dan gender data yang kosong akan diisi dengan modus.

```python
df['shpping_rate'].fillna('FREE Shipping',inplace=True)
df['gender'].fillna('men',inplace=True)
```

```python
df.isna().sum()
```

Selanjutnya akan dibuat kolom baru untuk memperlihatkan disrkon dari setiap barang.

```python
df['Discount'] = (df['Original_price']-df['price'])/df['Original_price']*100
```

```python
df.groupby(['category'])['perc_of_unit_sold'].mean().sort_values()[::-1].plot(kind='bar')
plt.grid(True)
plt.xlabel('kategori', fontweight = 'bold', fontsize = 8)
plt.ylabel('rata-rata persentase penjualan', fontweight = 'bold', fontsize = 8)
plt.title('Rata-Rata Persentase Penjualan Berdasarkan Kategori', fontweight = 'bold', fontsize = 14)
plt.savefig('Rata-Rata-Persentase-Penjualan-Berdasarkan-Kategori.png')
```
